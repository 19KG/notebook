```python
import random
def mov_type(reason):
    print("用户非常喜欢类型为"+reason+'的影片，所以给他推荐了同类影片')

def mov_language(reason):
    print("看来"+reason+"片是你的最爱，推荐一部非常棒的电影给你哟")

def mov_area(reason):
    print("不要问我为什么，"+ reason+ "的电影是你的最爱！")

def mov_publish_time(reason):
    print("时间是你选择电影的首要因素，给你推荐"+reason+"的电影，希望你喜欢")

def mov_director(reason):
    print(reason + "执导的电影确实值得一看，这里给你推荐一部")

def mov_screen_writer(reason):
    print(reason + "编剧太牛了，说实话，我也喜欢" + reason +"的作品")

def mov_actor(reason):
    print(reason + "主演电影棒棒的")
    
def select_mode(rel, reason):
    if rel == '类型':
        mov_type(reason)
    elif rel == '主演':
        mov_actor(reason)
    elif rel == '编剧':
        mov_screen_writer(reason)
    elif rel == '语言':
        mov_language(reason)
    elif rel == '导演':
        mov_director(reason)
    elif rel == '发行国家':
        mov_area(reason)
    else:
        mov_publish_time(reason)

for k, v in rec_dict.items():
    reason = random.choice(v)
#     print(reason)
    if len(reason) == 5:
        select_mode(reason[1], reason[2])
    elif len(reason) == 9:
        relation = ['导演','编剧','主演','发行国家','语言']
        for i, re in enumerate(reason):
            if re in relation:
                select_mode(reason[i], reason[i+1])
                break
```
