```python
all_bad_uid_list=[]
finished_uid_list=[]
```

```python
duplicat_mid=df_m['mid'].copy()
for j in range(2999):
    duplicat_mid=duplicat_mid.append(df_m['mid'],ignore_index=True)
```

```python
temp_filelist=filelist[0:1000]
```

```python
new_uid_list=[]
bad_uid_list=[]
path='/data/ng/FFM/test3/'
newfile=open('/data/ng/FFM/all_val.ffm','w')
for item in temp_filelist:    
    temp_uid=re.findall(r"\d+?\d*",item)
    fo=open(path+item,'r')
    temp_line=fo.readlines()
    if(len(temp_line)==3000):
        for line in temp_line:
            newfile.write(line)
    elif len(temp_uid)>0:
        bad_uid_list.append(temp_uid[0])
    if(len(temp_uid)>0):
        new_uid_list.append(temp_uid[0])
newfile.close()

for uid in bad_uid_list:
    new_uid_list.remove(uid)
    all_bad_uid_list.append(uid)
finished_uid_list.append(new_uid_list)
```

```python
ffm_model = xl.create_ffm()
ffm_model.setTest("/data/ng/FFM/all_val.ffm")
ffm_model.setSigmoid()
ffm_model.predict("/data/ng/FFM/model2.out", "/data/ng/FFM/output3.txt")
m_score=pd.read_csv("/data/ng/FFM/output3.txt",'r',header=None,names=['score'])
m_score=pd.concat([duplicat_mid,m_score],axis=1)
m_score=m_score.dropna()

recommand_file=open("/data/ng/FFM/recommand_resul3.txt",'a')
for i in range(len(new_uid_list)):
    temp_df=m_score[i*3000:(i+1)*3000]
    temp_df=temp_df.sort_values(['score'],ascending=False)
    command=temp_df.iloc[0:50]
    command_list=command['mid'].tolist()
    for command_movie in command_list:
        temp=str(new_uid_list[i])+'\t'+str(command_movie)+'\n'
        recommand_file.write(temp)
recommand_file.close()
```
