```python
import urllib.request
import json
from time import sleep
import string#后面处理页要用到
from itertools import islice

def next_n_lines(file_opened, N):
    return [x.strip() for x in islice(file_opened, N)]

html = r'https://api.douban.com/v2/movie/subject/'

print ('begin...')

dest = open('./result.txt','w+')

with open('mid.txt', 'r') as f:
    i = 1
    j = 1
    for i in range(int(2867/20)+1):
        sleep(20)
        lines  = next_n_lines(f, 20)
        for line in lines:
            mid = line.strip()
            url = html + mid  
            try:
                #url = 'https://movie.douban.com/subject/10569151/'
                hjson = json.loads(urllib.request.urlopen(url).read())
                title = hjson['title']
                newline = str(j)+','+mid+','+title.strip()
                #dest.write(newline+'\n')
                print(newline)
                sleep(1)
                j+=1
            except Exception as e:
                print ('404')
            #处理json，具体返回样例参照豆瓣API即可，输出格式：  排行：电影中文名---英文名（年代）
            i+=1
            if i%90 ==0:
                print("the "+str(i)+"* 90th mid has been requested")

print('over!')
dest.close()
```
