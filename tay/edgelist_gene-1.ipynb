```python
import codecs
import pandas as pd
import json
```

```python
kg = pd.read_csv("/data/tay/kg.txt", sep='\t',encoding='utf-8')
```

```python
kg.head()
```

```python
actors = kg[kg.relation=='主演']
director = kg[kg.relation=='导演']
screenwriter = kg[kg.relation=='编剧']
type = kg[kg.relation=='类型']
country = kg[kg.relation=='发行国家']
year = kg[kg.relation=='发行年份']
language = kg[kg.relation=='语言']
```

```python
my_actor = actors[['head', 'tail']]
my_director = director[['head', 'tail']]
my_screenwriter = screenwriter[['head', 'tail']]
my_type = type[['head', 'tail']]
my_country = country[['head', 'tail']]
my_year = year[['head', 'tail']]
my_lang = language[['head', 'tail']]
my_actor['name'] = json.dumps({'name':'主演'},ensure_ascii=False)
my_director['name'] = json.dumps({'name':'导演'},ensure_ascii=False)
my_screenwriter['name'] = json.dumps({'name':'编剧'},ensure_ascii=False)
my_type['name'] = json.dumps({'name':'类型'},ensure_ascii=False)
my_country['name'] = json.dumps({'name':'发行国家'},ensure_ascii=False)
my_year['name'] = json.dumps({'name':'发行年份'},ensure_ascii=False)
my_lang['name'] = json.dumps({'name':'语言'},ensure_ascii=False)
```

```python
my_edgelist = pd.DataFrame(columns=['head','tail', 'name'])
```

```python
my_edgelist = my_edgelist.append(my_actor, ignore_index=True)
my_edgelist = my_edgelist.append(my_director, ignore_index=True)
my_edgelist = my_edgelist.append(my_screenwriter, ignore_index=True)
my_edgelist = my_edgelist.append(my_type, ignore_index=True)
my_edgelist = my_edgelist.append(my_country, ignore_index=True)
my_edgelist = my_edgelist.append(my_year, ignore_index=True)
my_edgelist = my_edgelist.append(my_lang, ignore_index=True)
```
