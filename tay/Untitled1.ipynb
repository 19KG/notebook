```python
import codecs
import pandas as pd
```

```python
mid2_new = {}
```

```python
with codecs.open("ripple_mid.txt", "r", "utf-8") as f:
    for line in f.readlines():
        line = line.strip()
        mid, new = line.split("\t")[0], line.split("\t")[1]
        # print(type(mid))
        mid2_new[mid] = new
```

```python
print(mid2_new['1306483'])
```

```python
def mid_map(x):
    x = str(x)
    if x in mid2_new:
        return mid2_new[x]
    else:
        return 0
fd = pd.read_csv("final_data1.csv")
fd['mid'] = fd['mid'].apply(lambda x: mid_map(x))
```

```python
fd = fd[~fd['mid'].isin([0])]
```
