```python
import pandas as pd
import pickle
import pprint
import re
import rdflib
```

```python
with open('k-graph.bin', 'rb') as f:
    g = pickle.load(f)
```

# 类别

```python
result_list=list()
# iii=40
with open('good_list.txt','r') as f:
    line = f.readline()
    while line:
#     while iii>0:
#         print(line)
#         iii-=1
        str_begin='SELECT * WHERE {<'+line.strip()+'> <http://ckb.org/ontology/#类别> ?c}'
        res=g.query(str_begin)
        for r in res:
#             pprint.pprint(r)
            l=str(r)
            l_temp=list()
            a1=re.split('rdflib.term.Literal\(\'',l) 
            a2=a1[1].replace('\'),)','')+'\n'
            a3=re.split(r'[ ><,\、\n]',a2)
            for _ in a3:
                if(len(_)>0):
                    l_temp.append(_)
            l_temp=list(set(l_temp))
            a4=l_temp[0]
            if(len(l_temp)>1):
                for i in range(1,len(l_temp)):
                    a4+='_'+l_temp[i]
#             print(a4)
            temp=line.strip()+'\t'+a4+'\n'
            result_list.append(temp)
        line=f.readline()
#         break
```
