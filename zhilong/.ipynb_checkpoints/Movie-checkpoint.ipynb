### Process each attributes

```python
name_pattern = re.compile('\((.+?)\);')
# year_pattern = re.compile('(\d{4})(.*)年')
# date_pattern = re.compile('(\d+)(\-|\.)(\d+)(\-|\.)?(\d+)?')
date_pattern = re.compile('(20|19|18)\d{2}')
tag_pattern = re.compile('(.+?)\([0-9]+\);')

def extract_name(s: Union[str, None]) -> Union[List[str], None]:
    if type(s) is str:
        m = name_pattern.findall(s)
    else:
        m = None
    return m

def process_categories(s: Union[str, None]) -> Union[str, List[str], None]:
    if type(s) is str:
        m = list(i for i in s.split(';') if len(i) > 0)
        if len(m) > 1:
            m = m[:-1]
    else:
        m = None
    return m

def process_country(s: Union[str, None]) -> Union[str, List[str], None]:
    if type(s) is str:
        m = list(i for i in s.split(';') if len(i) > 0)
    else:
        m = None
    return m

def process_release_time(s: Union[str, int, None]) -> Union[int, float]:
    if type(s) is str:
        res = date_pattern.search(convert_fullwidth_to_halfwidth(s))
        if res is not None:
            year = int(res[0])
        else:
            year = np.NaN
    else:
        year = np.NaN
    return year

def convert_fullwidth_to_halfwidth(s):
    return ''.join(list(chr(ord(ch) - 65248) if 65281 < ord(ch) < 65374 else ch for ch in s))

def process_tag(s: Union[str, None]) -> Union[List[str], None]:
    if type(s) is str:
        return tag_pattern.findall(s)
    else:
        return None
```

```python
data = source_data.copy(deep=True)
data['director'] = data['director'].map(extract_name)
data['scriptwriter'] = data['scriptwriter'].map(extract_name)
data['mainPlayers'] = data['mainPlayers'].map(extract_name)
data['categories'] = data['categories'].map(process_categories)
data['productCountry'] = data['productCountry'].map(process_country)
data['tags'] = data['tags'].map(process_tag)
data['releasedTime'] = data['releasedTime'].map(process_release_time)
```

```python
# Replace some specific names

date_replace_dict = {
    '隐瞒 buried truth': 2008,
    'The Verve the Video 96-98': 1999,
    '我想飞': 2008,
    '死尸档案 The Necro Files (1997)': 1997,
    'Space Zombie Bingo': 1993,
    '谁的孩子': 2010,
    '我们正在老去': 2011
}

for i in data.index:
    if type(data.at[i, 'releasedTime']) is str and data.at[i, 'name'] in date_replace_dict:
        data.at[i, 'releasedTime'] = date_replace_dict[data.at[i, 'name']]
```

## Construct Knowledge Graph with Movie.txt

```python
graph = Graph()

ont = Namespace('http://ckb.org/ontology/#')
PREFIX = 'http://ckb.org/ontology/#'

attr = ['director', 'scriptwriter', 'mainPlayers', 
        'categories', 'productCountry', 'duration', 
        'releasedTime', 'tags', 'language']

relation = ['导演', '编剧', '主演', '类型', '发行国家', '时长', '发行年份', '标签', '语言']
reverse_relation = ['执导', '剧作', '参演', '包含', '发行自', '上映', '标记电影', '包括']

for idx, row in data.iterrows():
    mid = str(row['mid'])
    if row['director'] is not None:
        for item in row['director']:
            graph.add((URIRef(PREFIX + mid), 
                   ont.导演, 
                   URIRef(PREFIX + item.replace(' ', '_'))))
            graph.add((URIRef(PREFIX + item.replace(' ', '_')), 
                   ont.执导, 
                   URIRef(PREFIX + mid)))
    if row['scriptwriter'] is not None:
        for item in row['scriptwriter']:
            graph.add((URIRef(PREFIX + mid), 
                   ont.编剧, 
                   URIRef(PREFIX + item.replace(' ', '_'))))
            graph.add((URIRef(PREFIX + item.replace(' ', '_')), 
                   ont.剧作, 
                   URIRef(PREFIX + mid)))
    if row['mainPlayers'] is not None:
        for item in row['mainPlayers']:
            graph.add((URIRef(PREFIX + mid), 
                   ont.主演, 
                   URIRef(PREFIX + item.replace(' ', '_'))))
            graph.add((URIRef(PREFIX + item.replace(' ', '_')), 
                   ont.参演, 
                   URIRef(PREFIX + mid)))
    if row['categories'] is not None:
        for item in row['categories']:
            graph.add((URIRef(PREFIX + mid), 
                   ont.类型, 
                   Literal(item.replace(' ', '_'))))
            graph.add((Literal(item.replace(' ', '_')), 
                   ont.包含, 
                   URIRef(PREFIX + mid)))
    if row['productCountry'] is not None:
        for item in row['productCountry']:
            graph.add((URIRef(PREFIX + mid), 
                  ont.发行国家, 
                  URIRef(PREFIX + item.replace(' ', '_'))))
            graph.add((URIRef(PREFIX + item.replace(' ', '_')), 
                   ont.发行自, 
                   URIRef(PREFIX + mid)))
    if not np.isnan(row['duration']):
        graph.add((URIRef(PREFIX + mid), 
              ont.时长, 
              Literal(str(int(row['duration'])))))
    if not np.isnan(row['releasedTime']):
        graph.add((URIRef(PREFIX + mid), 
              ont.发行年份, 
              Literal(str(int(row['releasedTime'])))))
#     if row['tags'] is not None:
#         for item in row['tags']:
#             g.add((URIRef(PREFIX + mid), 
#                   ont.标签, 
#                   URIRef(PREFIX + item.replace(' ', '_'))))
#             g.add((URIRef(PREFIX + item.replace(' ', '_')), 
#                   ont.标记电影, 
#                   URIRef(PREFIX + mid)))
    if type(row['language']) is str:
        graph.add((URIRef(PREFIX + mid), 
             ont.语言, 
             URIRef(PREFIX + item.replace(' ', '_'))))
        graph.add((URIRef(PREFIX + item.replace(' ', '_')), 
               ont.包括,
              URIRef(PREFIX + mid)))
```
