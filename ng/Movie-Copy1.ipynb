```python
from pyparsing import ParseException
import pprint
import pandas as pd

QUERY = r'SELECT * WHERE {<http://ckb.org/ontology/#_%s_> ?b ?c}'

c_list = list()
e_list = list()

data = pd.read_csv("C:/Users/Thesharing/Downloads/YunPan/DeeCamp_2018/Movie.txt", sep='\t', encoding='utf-8')

for i, name in enumerate(data['name']):
    if name is not None and type(name) is str:
        n = name.split(' ')[0]
        try:
            res = g.query(QUERY % n)
            if len(res) > 0:
                c_list.append(n)
        except ParseException:
            print('Error at %d %s %s' % (i, name, n))
            e_list.append(n)
```

```python
import re

pattern = re.compile('([A-Za-z])+')

good_list = list()
for item in c_list:
    m = pattern.match(item)
    if m is None or len(m.group()) != len(item):
        good_list.append(item)
```

```python
import pandas as pd

data = pd.read_csv("C:/Users/Thesharing/Downloads/YunPan/DeeCamp_2018/Movie.txt", sep='\t', encoding='utf-8')
```

```python
import re
from typing import Union, List

def extract_name(s: Union[str, None]) -> Union[List[str], None]:
    if type(s) is str:
        m = re.findall('\((.+?)\);', s)
    else:
        m = None
    return m
```

```python
data['director'] = data['director'].map(extract_name)
data['scriptwriter'] = data['scriptwriter'].map(extract_name)
data['mainPlayers'] = data['mainPlayers'].map(extract_name)
```
